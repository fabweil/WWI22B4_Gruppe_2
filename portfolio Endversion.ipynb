```python
#Die Daten der CSV-Datei werden geladen
daten = pd.read_csv("Final_Augmented_dataset_Diseases_and_Symptoms.csv")
```

```python
#Das Dataframe wird in ein anderes kopiert und dort nach den diseases gruppiert
daten_gruppiert = daten.groupby("diseases").sum()
print(daten_gruppiert)
```

```python
#Diese Funktion soll später überprüfen ob der Schwellenwert 20% überschritten wurde. In dem Fall wird der Wert durch 1 ersetzt. Sonst wird er zu null. 
def check_threshold(s):
    threshold = 0.2 # 20% Schwellenwert
    count_ones = s.sum()
    count_total = len(s)
    if count_ones / count_total >= threshold:
        return 1
    else:
        return 0
```

```python
#Die Funktion wird nun an dem Dataframe angewendet und in einem neuen Dataframe gespeichert
sz = daten.groupby('diseases').apply(lambda x: x.iloc[:,1:].apply(check_threshold)).reset_index()
```

```python
#Von jeder Spalte wird die Summe berechnet. Es gibt noch Spalten, die immer den Wert null haben.
summen = daten.sum()

spalten_mit_summe_null = summen[summen == 0]

print(spalten_mit_summe_null.index)
```

```python
#Das zweite Dataframe wird zu einer weiteren CSV-Datei
sz.to_csv("diseases_new.csv", encoding="utf-8", index=False, sep=",")
```

```python
#Die Spalten, die immer null sind, werden aus dem ursprünglichen Dataframe entfernt
daten = daten.drop(columns=spalten_mit_summe_null.index)
```

```python
#Diese Spalten, die immer null werden auch aus dem zweiten Dataframe entfernt
sz = sz.drop(columns=spalten_mit_summe_null.index)
```

```python
#Da die Daten immer noch mehr als 200 Spalten hat, müssen die meisten entfernt werden, damit nur noch etwa 60 übrig sind
#Dafür wird von jeder Spalte die Summe berechnet und dann werden die 270 Spalten mit der niedrigsten Summe entfernt
numerische_daten = daten.select_dtypes(include='number')

summen = numerische_daten.sum()

top_spalten = summen.sort_values(ascending=True)[:270]

zu_entfernende_spalten = top_spalten.index

daten = daten.drop(columns=zu_entfernende_spalten)
sz = sz.drop(columns=zu_entfernende_spalten)

print(daten)
```

```python
ursprüngliche_daten = daten.copy()
sz_original = sz.copy()
daten = daten.iloc[:,1:]
```

```python
spalten_namen = daten.columns.to_numpy()
```

```python
#Im Rahmen dieses Codeteils soll der Nutzer für jede Spalte eintippen, ob er dieses Symptom hat oder nicht
eingaben = []
for spalten_name in spalten_namen:
    eingabe = input(spalten_name + ": ");
    eingabe = eingabe.lower()
    while eingabe not in ["nein", "ja", "j", "n"]:
        print("Ungültige Eingabe! Nur Ja oder Nein sind zulässig")
        eingabe = input(spalten_name + ": ")
        
    if(eingabe=="ja" or eingabe=="j"):
        eingaben.append(1)
    else:
        eingaben.append(0)
```

```python
#Alle Krankheiten, werden in einem Array gespeichert
krankheiten = sz.iloc[:, 0].values[1:]
krankheiten
```

```python
#Hier wird überprüft inwiweit die eingegebenen Daten mit den jeweiligen Krankheiten aus dem sz-Dataframe übereinstimmen. 
#Die Krankheit mit der höchsten Übereinstimmung wird ausgegeben
ergebnisse = {}

for index, reihe in sz.iterrows():
    krankheit = reihe[0]
    symptome = reihe[1:]
    
    uebereinstimmung = (eingaben == symptome).mean()  # Prozentsatz der Übereinstimmung
    
    ergebnisse[krankheit] = uebereinstimmung

beste_krankheit = max(ergebnisse, key=ergebnisse.get)
uebereinstimmung_prozent = ergebnisse[beste_krankheit] * 100

print(f"Die beste Übereinstimmung ist bei Krankheit '{beste_krankheit}' mit {uebereinstimmung_prozent:.2f}% Übereinstimmung.")
```
