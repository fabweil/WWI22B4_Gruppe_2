```python
daten = pd.read_csv("Final_Augmented_dataset_Diseases_and_Symptoms.csv")
```

```python
daten_gruppiert = daten.groupby("diseases").sum()
print(daten_gruppiert)
```

```python
def check_threshold(s):
    threshold = 0.2 # 30% Schwellenwert
    count_ones = s.sum()
    count_total = len(s)
    if count_ones / count_total >= threshold:
        return 1
    else:
        return 0
```

```python
sz = daten.groupby('diseases').apply(lambda x: x.iloc[:,1:].apply(check_threshold)).reset_index()
```

```python
ergebnisse = {}

for index, reihe in sz.iterrows():
    krankheit = reihe[0]
    werte = reihe[1:]
    
    # Eingabe fÃ¼r diese Krankheit erhalten
    eingabe_fuer_krankheit = eingaben[index - 1]  # Beachte: Index - 1, da die erste Zeile ignoriert wird
    
    # Vergleiche die Eingabe mit den Werten dieser Zeile
    uebereinstimmung = (eingabe_fuer_krankheit == werte).mean()  # Prozentsatz der Ãœbereinstimmung
    
    ergebnisse[krankheit] = uebereinstimmung

# Ausgabe der Ergebnisse
for krankheit, uebereinstimmung in ergebnisse.items():
    print(f"Krankheit '{krankheit}': {uebereinstimmung * 100:.2f}% Ãœbereinstimmung")
    
```
